```python
import numpy as np # To handle image arrays
import pandas as pd # To handle input files
from sklearn.model_selection import train_test_split # for splitting data for training and test
from sklearn.ensemble import RandomForestClassifier # as name  implies
from sklearn.linear_model import SGDClassifier # as name  implies
from sklearn.svm import LinearSVC # as name  implies
from sklearn.neighbors import KNeighborsClassifier # as name  implies
from sklearn.metrics import accuracy_score
```

```python
# This train.csv has been downloaded from kaggle site
# https://www.kaggle.com/c/digit-recognizer

train = pd.read_csv("train.csv")
```

```python
train
```

```python
print(type(train))
```

```python
features = train.columns[1:]
```

```python
features
```

```python
x = train[features]
```

```python
x
```

```python
y = train['label']
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x/255., y, test_size=0.3, random_state=0)
```

```python
x_train
```

```python
y_train
```

# Random Forest

```python
clf_rf = RandomForestClassifier()
```

```python
clf_rf.fit(x_train, y_train)
```

```python
y_pred_rf = clf_rf.predict(x_test)
```

```python
acc_rf = accuracy_score(y_test, y_pred_rf)
```

```python
print ("RandomForectClassifier Accuracy: ", acc_rf)
```

# Stochastic Gradient Descent

```python
clf_sgd = SGDClassifier()
```

```python
clf_sgd.fit(x_train, y_train)
```

```python
y_preg_sgd = clf_sgd.predict(x_test)
```

```python
acc_sgd = accuracy_score(y_test, y_preg_sgd)
print("Stochastic Gradient Descent Accuracy: ", acc_sgd)
```

# Support Vector Machine

```python
clf_svm = LinearSVC()
clf_svm.fit(x_train, y_train)
```

```python
y_pred_svm = clf_svm.predict(x_test)
```

```python
acc_svm = accuracy_score(y_test, y_pred_svm)
print("SVM Accuracy: ", acc_svm)
```

# Nearest Neighbors

```python
clf_knn = KNeighborsClassifier()
clf_knn.fit(x_train, y_train)
```

```python
y_pred_knn = clf_knn.predict(x_test)
```

```python
acc_knn = accuracy_score(y_test, y_pred_knn)
print("KNN Accuracy: ", acc_knn)
```
