# Downloading JSON using Web-API and handling it

- from any-api.com, I obtained this xkcd URL for JSON file download

## JSON downlaod and image display
- XKCD

```python
import json
from urllib.request import urlopen
```

```python
# Get current comic from XKCD
with urlopen('http://xkcd.com/info.0.json') as response:
    source = response.read()
```

```python
current_comic = json.loads(source)
current_comic
```

```python
# Download the comic image
img_data = urlopen(current_comic['img']).read()
```

```python
# Read as image from raw bytes

from PIL import Image
import io
img = Image.open(io.BytesIO(img_data))
```

```python
%matplotlib inline

# Display the image
import matplotlib.pyplot as plt
plt.imshow(img);
```
